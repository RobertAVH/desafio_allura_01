### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
precio_t1 = sum(tienda['Precio'])
precio_t2 = sum(tienda2['Precio'])
precio_t3 = sum(tienda3['Precio'])
precio_t4 = sum(tienda4['Precio'])

print(f"El ingreso total de la tienda 1 es: ${precio_t1}")
print(f"El ingreso total de la tienda 2 es: ${precio_t2}")
print(f"El ingreso total de la tienda 3 es: ${precio_t3}")
print(f"El ingreso total de la tienda 4 es: ${precio_t4}")
```

# 2. Ventas por categoría

```python
def agrupar_contar_categoria(df):
  ventas_por_cat = df.groupby('Categoría del Producto').size()
  return ventas_por_cat.sort_values(ascending=False, inplace=False)

print(f"Ventas por categoria de la tienda 1: \n{agrupar_contar_categoria(tienda)}\n")
print(f"Ventas por categoria de la tienda 2: \n{agrupar_contar_categoria(tienda2)}\n")
print(f"Ventas por categoria de la tienda 3: \n{agrupar_contar_categoria(tienda3)}\n")
print(f"Ventas por categoria de la tienda 4: \n{agrupar_contar_categoria(tienda4)}\n")
```

# 3. Calificación promedio de la tienda


```python
def calcular_promedio_calificacion(df):
  return df['Calificación'].mean()

print(f"La calificación promedio de la tienda 1 es: {calcular_promedio_calificacion(tienda)}")
print(f"La calificación promedio de la tienda 2 es: {calcular_promedio_calificacion(tienda2)}")
print(f"La calificación promedio de la tienda 3 es: {calcular_promedio_calificacion(tienda3)}")
print(f"La calificación promedio de la tienda 4 es: {calcular_promedio_calificacion(tienda4)}")
```

# 4. Productos más y menos vendidos

```python
def mostrar_productos_mas_vendidos(df):
  return df.groupby('Producto').size().sort_values(ascending=False).head(5)

def mostrar_productos_menos_vendidos(df):
  return df.groupby('Producto').size().sort_values(ascending=False).tail(5)

print(f"Los 5 productos más vendidos de la tienda 1 son: \n{mostrar_productos_mas_vendidos(tienda)}\n")
print(f"Los 5 productos menos vendidos de la tienda 1 son: \n{mostrar_productos_menos_vendidos(tienda)}\n")

print(f"Los 5 productos más vendidos de la tienda 2 son: \n{mostrar_productos_mas_vendidos(tienda2)}\n")
print(f"Los 5 productos menos vendidos de la tienda 2 son: \n{mostrar_productos_menos_vendidos(tienda2)}\n")

print(f"Los 5 productos más vendidos de la tienda 3 son: \n{mostrar_productos_mas_vendidos(tienda3)}\n")
print(f"Los 5 productos menos vendidos de la tienda 3 son: \n{mostrar_productos_menos_vendidos(tienda3)}\n")

print(f"Los 5 productos más vendidos de la tienda 4 son: \n{mostrar_productos_mas_vendidos(tienda4)}\n")
print(f"Los 5 productos menos vendidos de la tienda 4 son: \n{mostrar_productos_menos_vendidos(tienda4)}\n")
```

# 5. Envío promedio por tienda

```python
def calcular_envio_promedio(df):
  return df['Costo de envío'].mean()

print(f"El costo de envío promedio de la tienda 1 es: {round(calcular_envio_promedio(tienda),2)}")
print(f"El costo de envío promedio de la tienda 2 es: {round(calcular_envio_promedio(tienda2),2)}")
print(f"El costo de envío promedio de la tienda 3 es: {round(calcular_envio_promedio(tienda3),2)}")
print(f"El costo de envío promedio de la tienda 4 es: {round(calcular_envio_promedio(tienda4),2)}")
```
